### content based movie recommdation system

```python

import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
movie = pd.read_csv('tmdb_5000_movies.csv')
credit = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movie.head()
```

```python
credit.head()
```

```python
# then now mearge this data 
movies = movie.merge(credit, on='title')
movies.head(1)
```

```python
# feature extreaction 
movies.info()
```

```python
movies = movies[['genres', 'id', 'keywords', 'overview', 'popularity', 'revenue', 'vote_average', 'vote_count', 'cast','title']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
import ast
ast.literal_eval

def convart(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(convart)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convart)
```

```python
import ast
ast.literal_eval

def convart2(obj):
    l = []
    counts = 0
    for i in ast.literal_eval(obj):
        if counts != 3:
            l.append(i['name'])
        else:
            break
    return l
```

```python
movies['cast'] = movies['cast'].apply(convart2)
movies.head()
```

```python
movies.info()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(1)
```

```python
def clean_text(x):
    if isinstance(x, list):
        return " ".join(x)
    elif isinstance(x, str):
        return x.replace('[','').replace(']','').replace("'", "").replace(",", " ")
    else:
        return ""
```

```python
# Apply this cleaning to each column before merging
movies['genres'] = movies['genres'].apply(clean_text)
movies['keywords'] = movies['keywords'].apply(clean_text)
movies['cast'] = movies['cast'].apply(clean_text)
movies['overview'] = movies['overview'].fillna('') 
```

```python
movies['tags'] = movies['genres'] + " " + movies['keywords'] + " " + movies['overview'] + " " + movies['cast']
```

```python
new_df = movies[['id','title','tags']]
```

```python
new_df
```

```python
new_df['tags']
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text) :
    y = []
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vactor = cv.fit_transform(new_df['tags']).toarray()
```

```python
vactor
```

```python
print(cv.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vactor)
```

```python
similarity[1]
```

```python
def recommender(movies):
    movies_index = new_df[new_df['title'] == movies].index[0]
    distance = similarity[movies_index]
    movies_list = sorted(list(enumerate(distance)), reverse=True, key=lambda x: x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommender("Spectre")
```
